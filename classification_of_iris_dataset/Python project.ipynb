# Classification Of Iris dataset using Machine Learning Algorithm in Python

```python

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python

dataset = pd.read_csv('iris.csv')
```

```python
dataset.head()
```

```python
%matplotlib inline
img=mpimg.imread('iris_types.jpg')
plt.figure(figsize=(20,40))

plt.imshow(img)
```

```python
X = dataset.iloc[:,:4].values
y = dataset['species'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 82)
```

```python
from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
nvclassifier.fit(X_train, y_train)
```

```python
sl=float(input("Enter Sepal Length:"))
sw=float(input("Enter Sepal width:"))
pl=float(input("Enter petal length:"))
pw=float(input("Enter petal width:"))
```

```python
y_pred = nvclassifier.predict([[sl,sw,pl,pw]])

print(y_pred)
if (y_pred=="setosa"):
    %matplotlib inline
    img=mpimg.imread('Setosa.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
elif (y_pred=="virginica"):
    %matplotlib inline
    img=mpimg.imread('Virginica.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
elif (y_pred=="versicolor"):
    %matplotlib inline
    img=mpimg.imread('Versicolor.jpg')
    plt.figure(figsize=(5,10))
    plt.imshow(img)
```
